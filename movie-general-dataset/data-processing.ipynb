```python
import pandas as pd
import numpy as np
```

```python
credits_df = pd.read_csv("film-data/credits.csv")
credits_df.head()
```

```python
movies_df = pd.read_csv('film-data/movies_metadata.csv')
ratings   = pd.read_csv('film-data/ratings.csv')
keywords  = pd.read_csv('film-data/keywords.csv')
```

### Формирую основной датафрейм из всех данных. Для этого делаю пару предобработок ниже

```python
ratings.rename(columns={'movieId':'id'}, inplace=True)
ratings.head()
```

```python
movies_df['id'].describe()
```

```python
movies_df.head()
false_idxs = movies_df['id'][movies_df['id'].str.contains('-')].index
movies_df.drop(false_idxs, inplace=True)

# меняю тип колонки id на целочисленный
movies_df['id'] = movies_df['id'].astype('int32')
```

```python
movies_df['id']
```

```python
df = movies_df.merge(credits_df, on='id')
df.head(3)
```

```python
df = df.merge(keywords, on='id')
df.head(3)
```

### Удаляю неважные колонки

```python
df.columns
```

```python
df.drop(['adult','belongs_to_collection','poster_path'], axis=1, inplace=True)
```

```python
df.info()
```

### Считаем пустоты

```python
df.isnull().sum()
```

Преобразовываю содержимое столбца ключевых слов (keywords)

```python
import ast
```

```python
def keywords_simplification(x):
    keyword_list = ast.literal_eval(x)
    cur_keywords = [key_dict['name'] for key_dict in keyword_list]
    return cur_keywords
```

```python
df["keywords"] = df["keywords"].apply(keywords_simplification)
```

Оставляю только значимые колонки

```python
processed_df = df[["id", "title", "overview", "popularity", "revenue", "vote_average", "vote_count", "keywords"]]
```

```python
processed_df.isnull().sum()
```

Так как объём данных умеренного количества, можно пренебречь обработкой и попыткой восстановления пустот (тем более что содержимое краткого описания фильма достоверно восстановить не удастся), и просто выкинуть сэмплы с пустыми данными

```python
processed_df.dropna(inplace=True)
```
